**Load driving video and source image**

```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage.transform import resize
from IPython.display import HTML
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
gpath = Path('/content/gdrive/My Drive/first-order-motion-model/')
```

Crop, resize, and cut videos: - https://ezgif.com/

```python
def display(source, driving, generated=None):
    fig = plt.figure(figsize=(8 + 4 * (generated is not None), 6))

    ims = []
    for i in range(len(driving)):
        cols = [source]
        cols.append(driving[i])
        if generated is not None:
            cols.append(generated[i])
        im = plt.imshow(np.concatenate(cols, axis=1), animated=True)
        # plt.axis('off')
        ims.append([im])

    ani = animation.ArtistAnimation(fig, ims, interval=37, repeat_delay=1000)
    plt.close()
    return ani
```

```python
source_image = imageio.imread(gpath/'sana_sq.jpg')
driving_video = imageio.mimread(gpath/'trumpsmug.mp4')

N = 300
#Resize image and video to NxN
source = resize(source_image, (N, N))[..., :3]
driving = [resize(frame, (N, N))[..., :3] for frame in driving_video]
```

```python
source_image = imageio.imread(gpath/'sana.jpg')
video_reader = imageio.get_reader(gpath/'trumpsmug.mp4')

source = resize(source_image, (256, 256))[..., :3]

print(f'Number of frames in video is {video_reader.get_length()}.')
print(f'Size of each frame is {np.shape(video_reader.get_data(0))}.')
```

```python
cnt = 0
video = []
for f in video_reader:
  
  video.append(resize(f, (256, 256))[..., :3])
  cnt += 1
  if cnt == 200: break
```

```python
HTML(display(source, driving).to_html5_video())
```

**Create a model and load checkpoints**

```python
from demo import load_checkpoints
generator, kp_detector = load_checkpoints(config_path='config/vox-256.yaml', 
                            checkpoint_path='/content/gdrive/My Drive/first-order-motion-model/vox-cpk.pth.tar')
```

**Perform image animation**

```python
from demo import make_animation
from skimage import img_as_ubyte

predictions = make_animation(source_image, driving_video, generator, kp_detector, relative=True)

#save resulting video
imageio.mimsave('../generated.mp4', [img_as_ubyte(frame) for frame in predictions])
#video can be downloaded from /content folder

HTML(display(source_image, driving_video, predictions).to_html5_video())
```

**In the cell above we use relative keypoint displacement to animate the objects. We can use absolute coordinates instead,  but in this way all the object proporions will be inherited from the driving video. For example Putin haircut will be extended to match Trump haircut.**

```python
predictions = make_animation(source_image, driving_video, generator, kp_detector, relative=False, adapt_movement_scale=True)
HTML(display(source_image, driving_video, predictions).to_html5_video())
```
